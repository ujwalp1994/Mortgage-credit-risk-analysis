##### Dropping some non nessarary correlations

```python
data_corr = data.drop(['BoEth', 'Race2', 'Race3', 'Race4', 'Race5','CoEth', 'Corace2', 'Corace3', 'Corace4', 'Corace5','Rent1', 'Rent2', 'Rent3', 'Rent4',
       'RentUt1', 'RentUt2', 'RentUt3', 'RentUt4'], axis=1)
```

```python
corr = data_corr.corr()
corr2 = corr[(corr >0.4) & (corr !=1)]
corr3 = corr2.dropna(thresh=1)
```

```python
from matplotlib import pyplot
pyplot.figure(figsize=(25, 18)) # width and height in inches
sns.heatmap(corr3, cbar = True,   annot=True,  cmap= 'coolwarm')
```
